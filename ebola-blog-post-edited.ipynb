This almost covers the entirety of our area of interest; it also includes some other surrounding countries from which, in theory, we'll filter out tweets from later.

It turns out that not one of the tweets returned from our query lay inside of our search radius (empty black circle) and instead they were seemingly randomly distributed around the continents. There were a few exceptions of focii that lay just outside our search radius (remember that the radius of each spot is not a reflection of any geographical measure and is simply proportional to its frequency oa appearance in the search results, so the intersection of our search radius and the Gambia marker doesn't mean anything since its focii doesn't intersect).  

We asked ourselves "whaaat?" and went back to the drawing board.  The results from our search were literally all noise.  We tried using the streaming api to get results from gps tagged tweets within a bounding box similar to our search radius, but it turns out that not a whole lot of people tweet from our area of interest, especially with gps tagged tweets.  In any event, we wanted to be able to utilize the extra information that users often put in their location status, so we went back to thinking about the REST-ful tweet search.

We reasoned that it would be interesting to investigate how our signal to noise ratio varied as we sized down our search radius.  If instead of one large search radius, what if we used a large amount of small search radii?  We hypothezied that at some point, we'd start getting different patterns of results from our original search.

To experiment, we picked out 7 of the most populated cities/towns in each of the three countries and then manually ran tweet searches at varying radii, starting usually from 50km.  What we noticed were large discontinuities in result patterns at region specific radii values.

Since this threshold was critical for us getting quality results for our project we dug in a little deeper. So, at "0" - i.e., the previously determined search radius threshold, result numbers drop quite fast and, for the most part, with a few interesting exceptions, quite homogeneously.  

This is interesting in itself, but for our purposes, it was necessary to know whether or not these result dropoff threshold result in actually more robust data itself, i.e., better signal to noise ratio.  We looked at two metrics:  the ratio of unique GPS tagged tweets returned to the unique total, and the ratio of unique "location hit" tweets returned to the unique total, whereby if a user has set her profile location, as discussed above, to any administrative area in the respective country of search, we consider that a signal.  

```python
fig2, axes2 = plt.subplots(nrows=2, ncols=3)
fig2.tight_layout()
_n = 0
for _country in [l_range, s_range, g_range]:
    for _city in _country.keys():
        data = list()
        for _key in ['th4', 'th3', 'th2', 'th1', 'th0', 'th-1', 'th-2', 'th-3', 'th-4']:
            try:
                r = float(_country[_city][_key]['num_gps_u'])/float(_country[_city][_key]['length_u'])
                data.append(r)
            except Exception, e:
                print _city
                print e
        axes2[0,_n].plot(data)
        axes2[0,_n].axis([0,8,0,1])
        axes2[0,_n].xaxis.set_ticklabels(["4", "3", "2", "1", "0", "-1", "-2", "-3", "-4"])
        hold = True
    _n+=1
    hold = False

_j = 0
for _country in [l_range, s_range, g_range]:
    for _city in _country.keys():
        if _city == "pass":
            continue
        else:
            data = list()
            for _key in ['th4', 'th3', 'th2', 'th1', 'th0', 'th-1', 'th-2', 'th-3', 'th-4']:
                try:
                    r = float(_country[_city][_key]['num_loc_u'])/float(_country[_city][_key]['length_u'])
                    data.append(r)
                except Exception, e:
                    print _city
                    print e
            axes2[1,_j].plot(data)
            axes2[1,_j].axis([0,8,0,1])
            axes2[1,_j].xaxis.set_ticklabels(["4", "3", "2", "1", "0", "-1", "-2", "-3", "-4"])
            hold = True
    _j+=1
    hold=False
axes2[0,0].set_title("Liberia GPS")
axes2[0,1].set_title("SierraLeone GPS")
axes2[0,2].set_title("Guinnea GPS")

axes2[1,0].set_title("Liberia LOC")
axes2[1,1].set_title("SierraLeone LOC")
axes2[1,2].set_title("Guinea LOC")
plt.show()        
```

The top three plots heavily infer that the proportion of gps tagged tweets soars pretty consistently at the radius threshold.  The location hit ratio plots on the bottom suggest mixed results; an intersting further study might be to investigate the differences in these data.

As a rough metric for the "goodness" of our data, we added both ratios together.  In general, this metric goes up at the threshold radius.  

```python
fig3, axes3 = plt.subplots(nrows=1, ncols=3)
fig3.tight_layout()
_n = 0
for _country in [l_range, s_range, g_range]:
    for _city in _country.keys():
        data = list()
        for _key in ['th4', 'th3', 'th2', 'th1', 'th0', 'th-1', 'th-2', 'th-3', 'th-4']:
            try:
                r = float(_country[_city][_key]['num_gps_u']+_country[_city][_key]['num_loc_u'])/float(_country[_city][_key]['length_u'])
                data.append(r)
            except Exception, e:
                print _key
                print e
        axes3[_n].plot(data)
        axes3[_n].xaxis.set_ticklabels(["4", "3", "2", "1", "0", "-1", "-2", "-3", "-4"])
        
        hold = True
    _n+=1
    hold = False


axes3[0].set_title("Liberia GPS+LOC")
axes3[1].set_title("SierraLeone GPS+LOC")
axes3[2].set_title("Guinea GPS+LOC")
```

ADDENDUM

Our maps and graphs did not appear on visual inspection to bear out our hypothesis that city population size was the reseason search radius size varied.  So we ran a quick statistical correlation to verify what the map seemed to show.  The result is a weak -.22, or otherwise stated a population only explains 22% of the variation in search radius size.  It is negative because the search radius gets bigger as the population gets smaller in our hypothesis.  

We also thought the number of results for a location could be indicative of the size of the user community.  It then might help explain the variation in search radius size but the correlation results were even lower -0.1 - the number of search results for a location only explains 10% of the variation in search radius size.  

It was at first tempting to hypothesize that the variation in search radius threshold might be a function of city population: smaller populations might be associated with smaller Twitter usage (gross and possibly per capita), and small towns may tend to be more geographically remote from highly populated areas, therefore limiting the number of available data resources that Twitter might turn to when choosing what algorithms or combination thereof are used to produce the returned tweets - i.e., the only available data is mostly that which is gps tagged within that radius.  

This turned out to not be the case - at least in all instances.  A quick calculation of correlations between population, threshold radius and number of tweets returned (above) don't suggest any kind of reliable trend, and the small sample size and large variations in the data limit the significance of those numbers, anyway.  However, it is worthwhile to note that the most populated cities (> 500,000) were each characterized by a radius threshold of 5km.  A more refined hypothesis might be, then, that the relationship between city population and search radius threshold may be highly correlated for large enough population sizes.  However, a more in-depth study is needed to confidently comment on the relationship between search radius size and query results outside of this regime, and in general, the variables that govern that interesting relationship.
