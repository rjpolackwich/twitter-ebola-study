```python
from twgeo import TwitterGeoSearch
```

```python
tw = TwitterGeoSearch()
```

```python
def show_geo_search_results(coords, s_radius):
    try:
        res = tw._get_users_from_tweet_search(geocode=coords+","+s_radius)['statuses']
    except Exception, e:
        print e
    for _t in res:
        print _t['user']['screen_name']
        if _t['user']['location'] is not None:
            print _t['user']['location']
        else:
            print "No Location"
        if _t['coordinates'] is not None:
            print _t['coordinates']['coordinates']
        else:
            print "No Coordinates"
        print "\n"
```

```python
show_geo_search_results(coords="8.883,-12.05", s_radius="25km")
```
